```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(11, 4)})
```

```python
# Load data from CSV
data = pd.read_csv("./out/protoshape/interim/pool_data_part_1.csv")
```

```python
data.head()
```

```python
data.plot
```

```python
cols_plot = ['Area', 'Mean', 'Radius']
data[cols_plot][data['Area'] != 0].plot()
```

```python
#data['Radius']=data['Radius']*10
cols_plot = ['Area']
ax = data[cols_plot][data['Area'] != 0].rolling(20).mean().plot(title='Area of melt pool')
ax.set_xlabel("Time (frames)")
ax.set_ylabel("Pixels")
```

```python
cols_plot = ['Radius']
ax = data[cols_plot][data['Area'] != 0].rolling(20).mean().plot(title='Radius of melt pool')
ax.set_xlabel("Time (frames)")
ax.set_ylabel("Pixels")
```

```python
cols_plot = ['Mean']
ax = data[cols_plot][data['Area'] != 0].rolling(20).mean().plot(title='Mean intensity of melt pool')
ax.set_xlabel("Time (frames)")
ax.set_ylabel("Intensity")
```

```python
data[cols_plot][data['Area'] != 0].ewm(span = 2000).mean().plot(style = 'b', label = ' Exponential moving average')
```
